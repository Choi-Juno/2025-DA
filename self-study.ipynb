```python
import csv

with open('data/seoul.csv', 'r', encoding="cp949") as f:
    data = csv.reader(f)
    next(data)
    result = []

    for row in data:
        if row[-1] != '':
            result.append(row[-1])
    print(len(result))
```

```python
import matplotlib.pyplot as plt

plt.plot(result, 'r')
plt.figure(figsize=(10, 2))
plt.show()
```

```python
with open('data/seoul.csv', 'r', encoding="cp949") as f:
    data = csv.reader(f)
    next(data)
    result = []

    for row in data:
        if row[-1] != '':
            if row[0].split('-')[1] =="08":
                result.append(float(row[-1]))

plt.plot(result, 'hotpink')
plt.show()
```

```python
with open('data/seoul.csv', 'r', encoding="cp949") as f:
    data = csv.reader(f)
    next(data)
    result = []

    for row in data:
        if row[-1] != '':
            if row[0].split('-')[1] =="06" and row[0].split('-')[2] == "03":
                result.append(float(row[-1]))

plt.plot(result, 'hotpink')
plt.show()
```

```python
with open('data/seoul.csv', 'r', encoding='cp949') as f:
    data = csv.reader(f)
    next(data)
    high = []
    low = []

    for row in data:
        if row[-1] != '' and row[-2] != '':
            if int(row[0].split('-')[0]) >= 1983:
                if row[0].split('-')[1] == '06' and row[0].split('-')[2] == '03':
                    high.append(float(row[-1]))
                    low.append(float(row[-2]))

plt.plot(high, 'hotpink', label="최고기온")
plt.plot(low, 'skyblue', label="최저기온")
plt.rc('font', family="AppleGothic")
plt.title("내 생일의 기온 변화 그래프")
plt.legend()
plt.show()
```
